```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import statistics
import statsmodels.api as sm
from scipy.stats import norm
from scipy.stats import t
```

```python
df1 = pd.read_excel('Данные для тестового задания.xlsx',sheet_name = 1)
```

```python
df1.head()
```

```python
for exp in [1,2,3]:
    data = df1[df1['experiment_num']==exp]
    test =data[data['experiment_group']=='test']['revenue']
    control = data[data['experiment_group']=='control']['revenue']
    test_mean = np.mean(test)
    control_mean= np.mean(control)
    
    test_std = np.std(test, ddof=1)
    control_std = np.std(control, ddof=1)

    test_size= len(test)
    control_size = len(control)

    dof = test_size + control_size - 2

    err_diff = np.sqrt(test_std**2/test_size +control_std**2/control_size)

    diff_mean= test_mean - control_mean
    t_stat = diff_mean / err_diff

    p_value = 2*(1-t.cdf(abs(t_stat), dof))

    print("Experiment",exp)
    print("t_stat=", t_stat)
    print("p_value=", p_value)
    print()
```
